```python
import pandas as pd
data = pd.read_csv('c:/Users/aqib farooq/Downloads/python daily tasks/student_data_clean.csv')
data
```

```python
#department k hisab se avg cgpa
pivot1 = pd.pivot_table(data,
                        values="CGPA",
                        index="Department",
                        aggfunc="mean")
print(pivot1)
```

```python
#dep or gender k hisab se avg cgpa
pivot2 = pd.pivot_table(data,
                        values="CGPA",
                        columns="Department",
                        index="Gender",
                        aggfunc="mean")
print(pivot2)
```

```python
#city or dep k hisab se avg study hours
pivot3 = pd.pivot_table(data, 
                        values="Study_Hours_Per_Day",
                        index="Department",
                        columns="City",
                        aggfunc="mean"
)
print(pivot3)
```

```python
#scolership student count
pivot4 = pd.pivot_table(data,
                        values="Student_ID",
                        index="Department",
                        columns="Scholarship",
                        aggfunc="count")
print(pivot4)
```
