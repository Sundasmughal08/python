```python
class employee:
    
    def __init__(self, name ,employee_id , salary):
        self.__name = name
        self.__id = employee_id
        self.__salary = salary
        self.__department = "general"
    
    # getter
    def get__name(self):
        return self.__name

    # setter
    def set_salary(self, new_salary):
        self.__salary = new_salary

    # bonus calculation
    def calculate__bonus(self):
        return self.__salary * 0.05
    
    # display method add karo
    def display_info(self):
        print(f"Name : {self.__name}")
        print(f"ID : {self.__id}")
        print(f"Salary : {self.__salary}")
        print(f"Department : {self.__department}")


# object creation
emp1 = employee("Ali", 101, 5000)

emp1.display_info()
print("Bonus:", emp1.calculate__bonus())
```

```python
class person:
    def __init__(self , name ):
        self.__name = name

    def get__name(self):
        return self.__name
    def set__name(self , new_name):
        self.__name = new_name

p1 = person("ali")
print("before change:" , p1.get__name())

p1.set__name("ahmad")
print("after change :", p1.get__name())
    
```

```python
class person:
    def __init__(self , name):
        self.__name = name

    def get__name(self):
        return self.__name
    

class student(person):
    def __init__(self, name, grade):
        super().__init__(name)
#constructtor call
        self.__grade = grade

    def get__details(self):
        return (f"name: {self.get__name()} , grade : {self.__grade}")
#object
s1 = student("ali", "a")
print(s1.get__details())
             
```
